```python
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import pandas as pd

# Set up dates
end = datetime.now()
start = datetime(end.year - 4, end.month, end.day)

# Define stock symbol for ONGC
stock = "ONGC.NS"  # Use ".NS" for Indian stocks on Yahoo Finance
ongc = yf.download(stock, start, end)

# Display basic information about the stock data
print(ongc.head())
print(ongc.describe())
print(ongc.shape)
print(ongc.info())
print(ongc.isna().sum())

# Function to plot graphs
def plot_graph(figsize, values, column_name):
    plt.figure(figsize=figsize)
    values.plot()
    plt.xlabel("years")
    plt.ylabel(column_name)
    plt.title(f"{column_name} of ONGC")
    plt.show()

# Plot data for all columns
for column in ongc.columns:
    plot_graph((15, 5), ongc[column], column)

# Calculate and plot a 250-day moving average
ongc['MA250'] = ongc['Close'].rolling(250).mean()
plot_graph((15, 5), ongc['MA250'], 'MA250')
```
